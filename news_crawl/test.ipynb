```python
import requests
from bs4 import BeautifulSoup

# URL 설정
# IT 분야 네이버 뉴스
url = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105"

# http 요청
headers = {"User-Agent":"Mozilla/5.0"}
response = requests.get(url, headers=headers)

# html 파싱
soup = BeautifulSoup(response.text, "html.parser")

# 뉴스 제목 가져오기
titles = soup.select(".sa_text_strong") #네이버 뉴스용

# 출력
if titles:
    print("\n=== 뉴스 제목 목록 ===")
    for idx, title in enumerate(titles, 1):
        print(f"{idx}. {title.text.strip()}")
else:
    print("\n❌ 뉴스 제목을 찾을 수 없습니다. CSS 선택자를 확인하세요.")
```

# 뉴스 제목 + 기사 링크

```python
import requests
from bs4 import BeautifulSoup

# 네이버 IT 뉴스 URL
url = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105"

# HTTP 요청
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)

# HTML 파싱
soup = BeautifulSoup(response.text, "html.parser")

# 뉴스 제목 가져오기 (sa_text_strong 클래스 사용)
titles = soup.select(".sa_text_strong")

# 출력
if titles:
    print("\n=== 뉴스 제목 목록 ===")
    for idx, title in enumerate(titles, 1):
        news_title = title.text.strip()
        news_link = title.find_parent("a")["href"]  # 상위 <a> 태그에서 링크 가져오기
        print(f"{idx}. {news_title} - {news_link}")
else:
    print("\n❌ 뉴스 제목을 찾을 수 없습니다. CSS 선택자를 확인하세요.")
```

# 뉴스 제목 + 링크 csv로 저장

```python
import requests
from bs4 import BeautifulSoup
import csv

# 네이버 IT 뉴스 URL
url = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105"

# HTTP 요청
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)

# HTML 파싱
soup = BeautifulSoup(response.text, "html.parser")

# 뉴스 제목 가져오기
titles = soup.select(".sa_text_strong")

# CSV 파일로 저장
with open("news.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["번호", "제목", "링크"])  # 헤더 추가

    for idx, title in enumerate(titles, 1):
        news_title = title.text.strip()
        news_link = title.find_parent("a")["href"]
        writer.writerow([idx, news_title, news_link])
        print(f"{idx}. {news_title} - {news_link}")  # 출력도 계속 유지

print("\n✅ 뉴스 데이터를 'news.csv' 파일로 저장 완료!")
```
